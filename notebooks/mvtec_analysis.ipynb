### functions

```python
def get_impaths(main_path):
  return sorted([f for f in glob.glob(main_path+'*.png', recursive = True)])

def get_datasets_names(main_path):
    return [name for name in os.listdir(main_path) if os.path.isdir(os.path.join(main_path, name))]

def get_impaths(main_path):
    return sorted([f for f in glob.glob(main_path+'*.png', recursive = True)])

def visualize_images():
    #bottiglie
    good = Image.open("/home/ubuntu/workspace/dataset/bottle/train/good/000.png")
    bad1 = Image.open("/home/ubuntu/workspace/dataset/bottle/test/broken_large/000.png")
    bad2 = Image.open("/home/ubuntu/workspace/dataset/bottle/test/broken_small/000.png")
    bad3 = Image.open("/home/ubuntu/workspace/dataset/bottle/test/contamination/000.png")
    fig, axs = plt.subplots(1, 4, figsize=(15,15))
    axs[0].set_title('normal')
    axs[0].imshow(good)
    axs[1].set_title('broken small')
    axs[1].imshow(bad1)
    axs[2].set_title('broken large')
    axs[2].imshow(bad2)
    axs[3].set_title('contamination')
    axs[3].imshow(bad3)
    
    
    #tiles
    good = Image.open("/home/ubuntu/workspace/dataset/tile/train/good/000.png")
    bad1 = Image.open("/home/ubuntu/workspace/dataset/tile/test/crack/000.png")
    bad2 = Image.open("/home/ubuntu/workspace/dataset/tile/test/glue_strip/000.png")
    bad3 = Image.open("/home/ubuntu/workspace/dataset/tile/test/gray_stroke/000.png")
    bad4 = Image.open("/home/ubuntu/workspace/dataset/tile/test/oil/000.png")
    bad5 = Image.open("/home/ubuntu/workspace/dataset/tile/test/rough/000.png")

    fig, axs = plt.subplots(1, 3, figsize=(15,15))
    axs[0].set_title('normal')
    axs[0].imshow(good)
    axs[1].set_title('crack')
    axs[1].imshow(bad1)
    axs[2].set_title('glue strip')
    axs[2].imshow(bad2)
    fig, axs = plt.subplots(1, 3, figsize=(15,15))
    axs[0].set_title('gray stroke')
    axs[0].imshow(bad3)
    axs[1].set_title('oil')
    axs[1].imshow(bad4)
    axs[2].set_title('rough')
    axs[2].imshow(bad5)
    
    
    #toothbrush
    good = Image.open("/home/ubuntu/workspace/dataset/toothbrush/train/good/000.png")
    bad1 = Image.open("/home/ubuntu/workspace/dataset/toothbrush/test/defective/000.png")

    fig, axs = plt.subplots(1, 2, figsize=(15,15))
    axs[0].set_title('normal')
    axs[0].imshow(good)
    axs[1].set_title('defective')
    axs[1].imshow(bad1)
    
    
    #screw
    good = Image.open("/home/ubuntu/workspace/dataset/screw/train/good/000.png")
    bad1 = Image.open("/home/ubuntu/workspace/dataset/screw/test/scratch_neck/000.png")
    bad2 = Image.open("/home/ubuntu/workspace/dataset/screw/test/manipulated_front/000.png")
    bad3 = Image.open("/home/ubuntu/workspace/dataset/screw/test/thread_side/000.png")
    bad4 = Image.open("/home/ubuntu/workspace/dataset/screw/test/thread_top/000.png")

    fig, axs = plt.subplots(1, 5, figsize=(15,15))
    axs[0].set_title('normal')
    axs[0].imshow(good)
    axs[1].set_title('scratch neck')
    axs[1].imshow(bad1)
    axs[2].set_title('manipulated front')
    axs[2].imshow(bad2)
    axs[3].set_title('thread side')
    axs[3].imshow(bad3)
    axs[4].set_title('thread top')
    axs[4].imshow(bad4)
```

## Available (local) datasets and counting images for each class

```python
main_path = '/home/ubuntu/workspace/dataset/'
sub_folders = get_datasets_names(main_path)
print('available datasets:', sub_folders)

for folder in sub_folders:
    train_path = main_path+folder+'/train/good/'
    x = len(get_impaths(train_path))
    print(str.upper(folder)+':', x, 'anomaly free images')
    
    test_path = main_path+folder+'/test/'
    classes = get_datasets_names(test_path)
    for data_class in classes:
        print('    '+data_class+':', len(get_impaths(test_path+data_class+'/')))
```
